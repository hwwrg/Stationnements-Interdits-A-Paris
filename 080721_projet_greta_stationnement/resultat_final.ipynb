```python
# Q2 : Quels types des stationnements interdits étaient plus / moins populaires
# dans arrondissements de Paris ?
```

```python
# Construire la matrice d'arrondissement et combien de fois on a un type de stationnement
freq = df.groupby(['Arrondissement', 'Regime_particulier']).size()
freq
```

```python
# déviser une colonne en 3 colonnes
freq = freq.reset_index()
freq
```

```python
type(freq)
```

```python
# Calculer les types des stationnements interdits étaient plus / moins populaires
groups = freq.groupby(by=['Arrondissement'])
# Les plus populaires
freq2=groups.apply(lambda g: g[g[0] == g[0].max()]) 
freq2
```

```python
# Renomer les colonnes
freq2 = freq2.rename(columns={'Regime_particulier':'Max_Regime_particulier', 0:'Numbre_Max'})
freq2
```

```python
# Refaire l'index et supprimer la colonne 'Arrondissement'
freq2 = freq2.reset_index(level=1, drop=True).drop('Arrondissement', axis='columns')
freq2
```

```python
# Les plus populaires
freq3=groups.apply(lambda g: g[g[0] == g[0].min()])
freq3
```

```python
# Renomer les colonnes
freq3 = freq3.rename(columns={'Regime_particulier':'Min_Regime_particulier', 0:'Numbre_Min'})
freq3
```

```python
# Refaire l'index et supprimer la colonne 'Arrondissement'
freq3 = freq3.reset_index(level=1, drop=True).drop('Arrondissement', axis='columns')
freq3
```

```python
# Merger 'freq2' et 'freq3'
df_Q2 = freq2.join(freq3, how='right')
df_Q2
```

```python
# Supprimer l'arrondissement '0'
df_Q2 = df_Q2.drop(0, axis='index')       # 0 sans ''
df_Q2
```

```python
df_Q2.to_csv('dfq2.csv')
```
